```python
import pandas as pd
import numpy as np

#df  = pd.read_csv('b_should_be_easy.in', sep=' ', header=None)
df  = pd.read_csv('c_no_hurry.in', sep=' ', names=['x_start', 'y_start', 'x_end', 'y_end', 'earliest', 'latest'], 
                                                         header=None, skiprows=1)

#bonus = 
distances = abs(df['x_start'] - df['x_end']) + abs(df['y_start'] - df['y_end']) 
times = df['latest'] - df['earliest']
hurriness = distances / times
distances.head()
```

<h3>Distribution of distances</h3> 

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 5))
plt.hist(distances, bins=20, normed=True, label='Distance')
#plt.hist(times, bins=20, normed=True, alpha=0.7, label='Waiting window')
plt.legend()
#plt.title('Label distribution over similarity', fontsize=15)
plt.xlabel('Distances', fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(15, 5))
plt.hist(hurriness, bins=40, normed=True, label='Distance')
plt.legend()
plt.xlabel('Hurriness', fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(8, 8))

plt.scatter(df['x_end'],   df['y_end'],   c='b', label ='end')
plt.scatter(df['x_start'], df['y_start'], c='r', label='start')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15, 5))
plt.hist(df['earliest'], bins=40, normed=True)
plt.xlabel('Starting time', fontsize=15)
plt.show()
```
